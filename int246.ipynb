```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
Fever=ctrl.Antecedent(np.arange(95,105,1), 'Fever')
MusclePain=ctrl.Antecedent(np.arange(0,11,1), 'MusclePain')
Vomiting=ctrl.Antecedent(np.arange(0,11,1),'Vomiting')
Rash=ctrl.Antecedent(np.arange(0,11,1),'Rash')
Symptoms_Periods=ctrl.Antecedent(np.arange(0,7,1),"Symptoms_Periods")
Dengue=ctrl.Consequent(np.arange(0,101, 1), 'Dengue')
```

```python
Fever['low']=fuzz.trimf(Fever.universe,[95,95,98])
Fever['medium']=fuzz.trimf(Fever.universe, [96,99,102])
Fever['high']=fuzz.trimf(Fever.universe, [100,102,104])
Fever.view()
```

```python
MusclePain['No']=fuzz.trimf(MusclePain.universe,[0,0,4])
MusclePain['Maybe']=fuzz.trimf(MusclePain.universe, [2,4,8])
MusclePain['Yes']=fuzz.trimf(MusclePain.universe, [6,10,10])
MusclePain.view()
```

```python
Vomiting['No']=fuzz.trimf(Vomiting.universe,[0,0,4])
Vomiting['Maybe']=fuzz.trimf(Vomiting.universe, [2,4,8])
Vomiting['Yes']=fuzz.trimf(Vomiting.universe, [6,10,10])
Vomiting.view()
```

```python
Rash['No']=fuzz.trimf(Rash.universe,[0,0,4])
Rash['Maybe']=fuzz.trimf(Rash.universe,[2,4,8])
Rash['Yes']=fuzz.trimf(Rash.universe,[6,10,10])
Rash.view()
```

```python
Symptoms_Periods['No']=fuzz.trimf(Symptoms_Periods.universe,[0,0,2])
Symptoms_Periods['Maybe']=fuzz.trimf(Symptoms_Periods.universe, [1,2,3])
Symptoms_Periods['Yes']=fuzz.trimf(Symptoms_Periods.universe, [2,4,6])
Symptoms_Periods.view()
```

```python
Dengue['No']=fuzz.trimf(Dengue.universe,[0,0,40])
Dengue['Maybe']=fuzz.trimf(Dengue.universe, [20,50,80])
Dengue['Yes']=fuzz.trimf(Dengue.universe, [ 60,100,100])
Dengue.view()
```

```python
rule1=ctrl.Rule(Fever['low']| MusclePain['No']| Vomiting['No']| Rash['No']| Symptoms_Periods['No'], Dengue['No'])
rule2=ctrl.Rule(Fever['low']| MusclePain['No']| Vomiting['No']| Rash['Maybe']| Symptoms_Periods['No'], Dengue['No'])
rule3=ctrl.Rule(Fever['low']| MusclePain['No']| Vomiting['No']| Rash['Yes']| Symptoms_Periods['No'], Dengue['No'])
rule4=ctrl.Rule(Fever['medium']| MusclePain['No']| Vomiting['No']| Rash['No']| Symptoms_Periods['Maybe'], Dengue['Maybe'])
rule5=ctrl.Rule(Fever['medium']| MusclePain['No']| Vomiting['No']| Rash['Maybe']| Symptoms_Periods['Maybe'], Dengue['Maybe'])
rule6=ctrl.Rule(Fever['medium']| MusclePain['No']| Vomiting['No']| Rash['Yes']| Symptoms_Periods['Maybe'], Dengue['Maybe'])
rule7=ctrl.Rule(Fever['medium']| MusclePain['No']| Vomiting['No']| Rash['Yes']| Symptoms_Periods['Maybe'], Dengue['Maybe'])
rule8=ctrl.Rule(Fever['high']| MusclePain['Yes']| Vomiting['No']| Rash['No']| Symptoms_Periods['Yes'], Dengue['Yes'])
rule9=ctrl.Rule(Fever['high']| MusclePain['Yes']| Vomiting['Yes']| Rash['Maybe']| Symptoms_Periods['Yes'], Dengue['Yes'])
rule10=ctrl.Rule(Fever['high']| MusclePain['Yes']| Vomiting['Yes']| Rash['Yes']| Symptoms_Periods['Yes'], Dengue['Yes'])
```

```python
Dengue_ctrl=ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10])
Occur=ctrl.ControlSystemSimulation(Dengue_ctrl)
Occur.input['Fever']=float(input("Enter the temperature "))
Occur.input['MusclePain']=float(input("Enter the value "))
Occur.input['Vomiting']=float(input("Enter the value1 "))
Occur.input['Rash']=float(input("Enter the value2 "))
Occur.input['Symptoms_Periods']=float(input("Enter the value3 "))
Occur.compute()
print(Occur.output['Dengue'])
Dengue.view(sim=Occur)
```
